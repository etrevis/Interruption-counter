```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import os

sns.set(style="whitegrid", palette='pastel')
```

```python
data = pd.read_csv('workfromhome_interruptions.csv')

data['Time'] = pd.to_datetime(data['Time'], format='%Y-%m-%d %H:%M:%S')

def get_hour(date):
    hours = (date.dt.hour + date.dt.minute/60)
    return hours.to_list()

data['Hour of the day'] = get_hour(data['Time'])

def is_self(a):
    if a==1:
        return 'Self-distracted'
    elif a==0:
        return 'Interrupted'
    
data['Type'] = pd.Series([is_self(i) for i in data['Interrupted'].to_list()])
data['Interruption density'] = ' '

data.describe()
```

```python
plot_ = sns.violinplot(x='Hour of the day', y = 'Interruption density', hue="Type", data = data,
                       split=True, palette='pastel', inner='sticks', scale = 'count',
                       scale_hue=False, bw=0.18)

plot_.set_xlim(6,22)
plot_.text(0.85, 0.85,'Text Here', fontsize=9)
#plot_.figure.set_size_inches(10, 6)
```

```python
for lol in range(10,len(data.index.to_list())+1):
    d = data[data.index < lol]

    fig, ax = plt.subplots()
    sns.violinplot(x='Hour of the day', y = 'Interruption density', hue="Type", data = d,
                           split=True, palette='pastel', inner='sticks', scale = 'count', ax= ax,
                           scale_hue=False, bw=0.25)

    ax.set_xlim(6,22)
    ax.text(6.3,0.45, ('Date: ' + str(data['Time'].dt.date.iloc[lol-1])), fontsize=9,
            bbox=dict(facecolor='white', edgecolor='gray', boxstyle='round'))

    plt.savefig(os.path.join(os.getcwd(),'Figs',str(lol)+'.png'),dpi=300, bbox_inches='tight')
    del fig, ax, d
    print(lol)

print('Done!')
```
